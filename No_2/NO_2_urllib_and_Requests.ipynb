# urllib的使用
是python3自带的内置模块
包含的方法：
- urllib.request:打开和读取URL
- urllib.error:包含urllib.request抛出的异常
- urllib.parse:用于解析URL
- urllib.robotparser:用于解析robots.txt

## urllib的网址请求

```python
import urllib.request
response = urllib.request.urlopen('https://cn.bing.com/?mkt=zh-CN')
print(response.read().decode('utf-8')) #调用read方法可以得到返回的网页内容并打印网页代码
```

如果网速比较慢或者所请求的网站打开较慢，可以设置一个超时限制，加入timeout参数

```python
import urllib.request
response = urllib.request.urlopen('https://cn.bing.com/?mkt=zh-CN',timeout=8)
print(response.read().decode('utf-8')) #使用read方法读取网页，decode设置编码方式
```

## 网页的保存和异常处理

```python
try:
    x = urllib.request.urlopen('https://cn.bing.com/?mkt=zh-CN')
    #print(x.read)
    save = open('1.html','w')
    save.write(str(x.read()))
    save.close()
except Exception as e:
    print(str(e))
```

## 构造请求对象Requests
使用Requests模仿浏览器，避免一些简单的反爬措施，"远古的方法"

```python
import urllib.request
request = urllib.request.Request('https://cn.bing.com/?mkt=zh-CN') # 用Request构造Request对象类
response = urllib.request.urlopen(request)
print(response.read().decode('utf-8'))
```

## SSL认证
有SSL认证的网页一般是安全的，如果遇到没有SSL认证的网页可以添加SSL忽略警告而继续访问

```python
import urllib
import urllib.request
# 对SSL进行设置，忽略警告，继续进行访问’
import ssl
ssl._create_default_https_context_ = ssl._create_unverified_context
url = 'https://www.12306.cn/mormhweb/'
response = urllib.request.urlopen(url=url)
print(response.read().decode('utf-8'))
```

这里也可以强制https认证（使用Requests请求）

```python
import requests
# urllib3官方强制认证https安全证书，用于解决警告
import urllib3
urllib3.disable_warnings()
url = 'https://www.12306.cn/mormhweb/'
# requests操作简单，False表示不要进行认证
response = requests.get(url=url,verify=False)
# 设置返回编码方式
response.encoding='utf-8'
print(response.text)
```

# Requests中post的使用
除了get请求，还有Post，post的作用是将要处理的数据提交给服务器  
优点如下：  
1.比GET更安全  
2.可以上传的数据更大，可以使用POST发送文本数据和二进制数据  
缺点如下：  
1.Post请求不会被缓存  
2.POST请求不会被保留在浏览器历史记录中  
3.POST发送的数据在URL中不可见，因此无法为特定查询的页面添加书签。  

## 提交数据表单

```python
import requests
data = {'name':'zyzyzyzy','age':'20'}
r = requests.post("https://httpbin.org/post",data=data)
print(r.text)
```

## 添加请求图

```python
import requests
import json
host = "https://httpbin.org/"
endpoint = "post"
url = ''.join([host,endpoint])
headers = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11'}
r = requests.post(url,headers=headers)
print(r.text)
```

## 提交json

```python
import requests
import json
host = "https://httpbin.org/"
endpoint = "post"
url = '/'.join([host, endpoint])
data = {
    "sites":[
        {"name":"zyzyzy","url":"https://www.icourse163.org/"},
        {"name":"pypypy","url":"https://space.bilibili.com/"},
        {"name":"fzufzu","url":"https://jwch.fzu.edu.cn/"}]}
r = requests.post(url,json=data)
if r.status_code == 200:
    try:
        response = r.json()
        print(response)
    except json.JSONDecodeError:
        print("解析JSON时出错。")
else:
    print(f"请求失败，状态码：{r.status_code}")
```

## URLError和HTTPError
URLError一般有几种情况：
- 没有网络连接
- 服务器连接失败
- 找不到指定服务器
- try except捕获异常

```python
import urllib
import urllib.request
url = "https://www.12306.cn/index/"
try:
    response = urllib.request.urlopen(url=url,timeout=5)
    print(response.read().decode('utf-8'))
except Exception as err:
    print(err)
```

HTTPError是在Web开发和网络请求中遇到的一种常见错误，通常由requests库或其他HTTP客户端库抛出，指示HTTP请求失败。HTTPError错误的发生通常与HTTP响应状态码有关，表示服务器返回了一个错误状态。以下是一些可能导致HTTPError的常见情况，分别对应不同的HTTP状态码：
1. 4xx 客户端错误
- 400 Bad Request：请求无效或无法被服务器理解。通常是发送的请求数据有误。
- 401 Unauthorized：请求未经授权。通常需要通过适当的验证头部进行身份验证。
- 403 Forbidden：服务器拒绝执行此请求。与401不同，403错误表示服务器理解请求但拒绝授权。
- 404 Not Found：服务器找不到请求的资源。最常见的情况，表示URL可能写错了。
- 405 Method Not Allowed：请求的方法（GET、POST、DELETE等）对指定的资源不适用。
2. 5xx 服务器错误
- 500 Internal Server Error：服务器遇到了一个未知的错误，无法完成客户端的请求。
- 502 Bad Gateway：作为网关或代理工作的服务器从上游服务器收到了无效响应。
- 503 Service Unavailable：服务器目前无法使用（由于超载或停机维护）。通常，这只是暂时状态。
- 504 Gateway Timeout：作为网关或代理的服务器未及时从上游服务器接收请求。
3. 其他情况
- 网络连接问题：客户端与服务器之间的网络连接出现问题，如DNS解析失败，网络中断，或连接超时，也可能导致请求失败，尽管这不一定总是抛出HTTPError。
- 重定向过多：客户端在尝试追踪重定向时超过了最大重定向次数，虽然这可能会引发TooManyRedirects异常，而不是HTTPError。

## 爬取翻译网站
[百度翻译](https://fanyi.baidu.com/)  
1. 输入翻译内容，检查网页
2. 发现翻译都在sug文件中（点预览）
3. 查看标头
<img src="fanyi.png" width='50%' height='50%'/>

点击“标头”可以发现请求网址：https://fanyi.baidu.com/sug ,请求方式：POST  
点击“响应”可以发现数据类型：字典，k：需要翻译的内容，v：翻译后的内容

```python
import json
import requests
url = 'https://fanyi.baidu.com/sug'
word = input("请输入要翻译的单词或短语：")
data = {'kw':word}
headers = {'User-Agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Mobile Safari/537.36 Edg/121.0.0.0'}
response = requests.post(url=url,data=data,headers=headers)
dic = response.json()
# print(dic)
filename = word+'.json'
with open(filename,'w',encoding='utf-8') as fp:
    json.dump(dic,fp=fp,ensure_ascii=False)
if dic['data']:
    j = dic['data'][0]['v']
    print(j)
else:
    print('数据列表为空或不存在该索引')
```
